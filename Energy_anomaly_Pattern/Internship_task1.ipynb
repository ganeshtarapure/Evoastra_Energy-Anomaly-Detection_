```python
 # Convert Timestamp to Datetime & Sort
df['timestamp'] = pd.to_datetime(df['timestamp'])
df = df.sort_values('timestamp')

print("Timestamp converted & sorted ✅")
```

```python
# Handle Missing Values
df = df.fillna(method='ffill')  # forward fill
df = df.fillna(method='bfill')  # backward fill

print("Missing values handled ✅")
```

```python
df.isnull().sum()
```

```python
# Cap Outliers (1%-99% Quantile)

# Explanation: Extreme values data ko distort kar sakte hain. 1% lowest aur 1% highest values cap kar dete hain.

for col in df.columns:
    if col != 'timestamp':
        lower = df[col].quantile(0.01)
        upper = df[col].quantile(0.99)
        df[col] = df[col].clip(lower, upper)

print("Outliers capped ✅")
```

```python
# Normalize Numeric
numeric_cols = df.columns.drop('timestamp')  # all numeric columns except timestamp

scaler = MinMaxScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

print("Normalization done ✅")

# Explanation: ML algorithms ke liye numeric values ko 0–1 range me lana zaruri hai.
```

```python
# feature engineering

# Rolling Statistics
# Rolling mean & rolling standard deviation (7-day window = 168 hours)
df['electricity_rolling_mean_168'] = df['electricity'].rolling(window=168).mean()
df['electricity_rolling_std_168'] = df['electricity'].rolling(window=168).std()
df['electricity_rolling_min_168'] = df['electricity'].rolling(window=168).min()
df['electricity_rolling_max_168'] = df['electricity'].rolling(window=168).max()

df[['electricity_rolling_mean_168','electricity_rolling_std_168']].tail()

# Rolling mean = pichle 7 din ka average, jo current value ko smooth karta hai.
# Rolling std = pichle 7 din ka variation, anomaly detect karne me madad karta hai.
```

```python
# Short-Term Rolling (24 Hours)
df['electricity_rolling_mean_24'] = df['electricity'].rolling(window=24).mean()
df['electricity_rolling_std_24'] = df['electricity'].rolling(window=24).std()

df[['electricity_rolling_mean_24','electricity_rolling_std_24']].tail()
```

```python
# Deviation from baseline
#Purpose: Current value ka comparison past trend se
df['electricity_deviation_168'] = (
    df['electricity'] - df['electricity_rolling_mean_168']
) / (df['electricity_rolling_std_168'] + 1e-5)

df['electricity_deviation_24'] = (
    df['electricity'] - df['electricity_rolling_mean_24']
) / (df['electricity_rolling_std_24'] + 1e-5)

df[['electricity_deviation_168','electricity_deviation_24']].tail()

#Deviation = (current - average) / std.
#Agar deviation zyada hai → unusual spike ya drop → anomaly detect kar sakte hain.
```

```python
# Time-based features
df['hour'] = df['timestamp'].dt.hour
df['day_of_week'] = df['timestamp'].dt.dayofweek
df['day_of_month'] = df['timestamp'].dt.day
df['month'] = df['timestamp'].dt.month
df['week_of_year'] = df['timestamp'].dt.isocalendar().week.astype(int)
df['quarter'] = df['timestamp'].dt.quarter
df['is_weekend'] = df['day_of_week'].isin([5, 6]).astype(int)

df[['hour','day_of_week','month','is_weekend']].head()
#Purpose: Seasonality aur time-of-day/week pattern capture karna.
```

```python
# Cyclical Encoding (Very Important)
df['hour_sin'] = np.sin(2 * np.pi * df['hour'] / 24)
df['hour_cos'] = np.cos(2 * np.pi * df['hour'] / 24)

df['month_sin'] = np.sin(2 * np.pi * df['month'] / 12)
df['month_cos'] = np.cos(2 * np.pi * df['month'] / 12)

df[['hour_sin','hour_cos','month_sin','month_cos']].head()
```

```python
# Lag Features
df['electricity_lag1'] = df['electricity'].shift(1)
df['electricity_lag3'] = df['electricity'].shift(3)
df['electricity_lag6'] = df['electricity'].shift(6)
df['electricity_lag24'] = df['electricity'].shift(24)
df['electricity_lag168'] = df['electricity'].shift(168)

df[['electricity_lag1','electricity_lag24']].tail()

# ag features → previous values ka reference, ML model future prediction ya anomaly detection me help karta hai.
```

```python
# Consumption Ratios
df['ratio_to_24hr_mean'] = df['electricity'] / (df['electricity_rolling_mean_24'] + 1e-5)
df['ratio_to_7day_mean'] = df['electricity'] / (df['electricity_rolling_mean_168'] + 1e-5)

df[['ratio_to_24hr_mean','ratio_to_7day_mean']].tail()
```

```python
# Difference Features
df['diff_1hr'] = df['electricity'] - df['electricity_lag1']
df['diff_24hr'] = df['electricity'] - df['electricity_lag24']

df[['diff_1hr','diff_24hr']].tail()
```

```python
# Expanding Statistics
df['expanding_mean'] = df['electricity'].expanding().mean()
df['expanding_std'] = df['electricity'].expanding().std()

df[['expanding_mean','expanding_std']].tail()
```
